Go to quandl.com, look for a dataset you want, and find its ID. For example, I searched for "Inflation", click on the Central Bank of Brazil Statistical Database, and found the National Consumer Price Index. The code for the dataset is "BCB/188". The units are 'Monthly % Variation', or the monthly change in the price index. Note how there was crazy hyperinflation in 1990, then some problematically high inflation until about 1995, after which inflation has held stable.

They actually replaced their entire currency at that point! They went from using the cruzeiro, which had faced so much inflation that it ceased functioning as a currency, to the real. The *Plano Real* was a series of strategic steps to limit inflation.

They changed currencies. They contracted fiscal and expansionary policies to lower government expenses. The opened the new currency at a high value relative to the US dollar, which gave the Brazilian government the upper hand in Balance of Payments. They raised interest rates, which attracted foreign capital and was actually able to cover the current government account deficits and build some international reserves. It was a huge success. Really incredible.

```python
# because we have only one columns, the .squeeze() method turns this DataFrame into a series.
# in general, np.squeeze() removes axes of length 1
brazil_ncpi = quandl.get('BCB/188', api_key=quandl_api_key).squeeze()
```

```python
sns.set_style('whitegrid')
brazil_ncpi.plot(figsize=(16,8), title='Brazil - National Consumer Price Index', ylabel='CPI, Monthly % Variation');
```
