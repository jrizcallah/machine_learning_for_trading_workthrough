## Setup the Logbook Stuff

```python
format_string = '[{record.time: %H:%M:%S.%f}]: {record.level_name}: {record.message}'

zipline_logging = NestedSetup([NullHandler(level=DEBUG),
                              StreamHandler(sys.stdout,
                                            format_string=format_string,
                                            level=INFO),
                              StreamHandler(sys.stderr, level=ERROR)])
zipline_logging.push_application()
log = Logger('Algorithm')
```

## Algorithm Setup

```python
MONTH = 21
YEAR = 12 * MONTH
N_LONGS = 50
N_SHORTS = 50
VOLUME_SCREEN = 500
```

```python
start = pd.Timestamp('2013-01-01', tz=UTC)
end = pd.Timestamp('2017-01-01', tz=UTC)
capital_base = 1e7
```

## Mean Reversion Factor

```python
class MeanReversion(CustomFactor):
    """ Object for computing the ratio of latest 
    monthly return to 12-month average,
    normalized by standard deviation of monthly returns"""
    inputs = [Returns(window_length=MONTH)]
    window_length = YEAR
    
    def compute(self, today, assets, out, monthly_returns):
        """Actually computes"""
        df = pd.DataFrame(monthly_returns)
        out[:] = df.iloc[-1].sub(df.mean()).div(df.std())
```

## Pipeline Setup

```python
def compute_factors():
    """ Creates factor pipeline, including MeanReversion;
    filter to include only the top companies by 30-day volume;
    get factor ranks"""
    mean_reversion = MeanReversion()
    dollar_volume = AverageDollarVolume(window_length=30)
    return Pipeline(columns = {'longs': mean_reversion.bottom(N_LONGS),
                               'shorts': mean_reversion.top(N_SHORTS),
                               'ranking': mean_reversion.rank(ascending=False)},
                    screen = dollar_volume.top(VOLUME_SCREEN))
```

```python
# Execute the pipeline and record the results, 
#    every day before trading opens
def before_trading_start(context, data):
    """ Run the factor pipeline """
    context.factor_data = pipeline_output('factor_pipeline')
    record(factor_data = context.factor_data.ranking)
    assets = context.factor_data.index
    record(prices = data.current(assets, 'price'))
```

## Rebalancing

```python
def rebalance(context, data):
    """ Compute new target long, short holdings;
    then place trade orders """
    factor_data = context.factor_data
    assets = factor_data.index
    
    longs = assets[factor_data.longs]
    shorts = assets[factor_data.shorts]
    
    divest = context.portfolio.positions.keys() - longs.union(shorts)
    
    log.info('{} | Longs: {:2.0f} | Shorts: {:2.0f} | {:,.2f}'.format(get_datetime().date(),
                                                                     len(longs),
                                                                     len(shorts),
                                                                     context.portfolio.portfolio_value))
    
    # divest first
    execute_trades(data, assets=divest, target_percent=0)
    # now long trades
    execute_trades(data, assets=longs, target_percent= 1/N_LONGS if N_LONGS else 0)
    execute_trades(data, assets=shorts, target_percent = 1/N_SHORTS if N_SHORTS else 0)
    
```

```python
def execute_trades(data, assets, target_percent):
    """ Place orders """
    for asset in assets:
        if data.can_trade(asset) and not get_open_orders(asset):
            order_target_percent(asset, target_percent)
```

## Initialize the Backtest

```python
def initialize(context):
    """ Create pipeline, schedule rebalancing,
    and set trading parameters"""
    attach_pipeline(compute_factors(), 'factor_pipeline')
    schedule_function(rebalance,
                      date_rules.week_start(),
                      time_rules.market_open(),
                      calendar=calendars.US_EQUITIES)
    
    set_commission(us_equities = commission.PerShare(cost=0.00075,
                                                     min_trade_cost=0.01))
    
    set_slippage(us_equities = slippage.VolumeShareSlippage(volume_limit=0.0025,
                                                            price_impact=0.01))
    
    
```

## Run the Algorithm

```python
backtest = run_algorithm(start=start,
                         end=end,
                         initialize=initialize,
                         before_trading_start=before_trading_start,
                         bundle='quandl',
                         capital_base=capital_base)
```

## Extract pyfolio Inputs

```python
returns, positions, transactions = extract_rets_pos_txn_from_zipline(backtest)
```

```python
# Save results
with pd.HDFStore('backtest.h5') as store:
    store.put('backtest/equal_weight', backtest)
    store.put('returns/equal_weight', returns)
    store.put('positions/equal_weight', positions)
    store.put('transactions/equal_weight', transactions)
```

## Plot Results

```python
fig, axes = plt.subplots(nrows=2, figsize=(20,8))

returns.add(1).cumprod().sub(1).plot(ax=axes[0], 
                                     title='Cumulative Returns')

transactions.groupby(transactions.dt.dt.day).txn_dollars.sum().cumsum().plot(ax=axes[1],
                                                                             title='Cumulative Transactions')

fig.tight_layout()
sns.despine();
```
